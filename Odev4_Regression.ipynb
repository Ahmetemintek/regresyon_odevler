```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn import linear_model
import warnings 
warnings.filterwarnings("ignore")
```

```python
df= pd.read_csv("Desktop/Bootcamp Ödevler/weatherHistory.csv")
df.head()
```

```python
# hedef değişken görünür sıcaklık ve sıcaklık arasındaki fark
#  Açıklayıcı değişkenler olarak nem ve rüzgar hızı
df["hedef"]= df["Temperature (C)"]-df["Apparent Temperature (C)"]
```

```python
df2= df.copy()
ozellikler= ["hedef","Temperature (C)", "Apparent Temperature (C)", "Humidity", "Wind Speed (km/h)"]
for i in df.columns:
    if i not in ozellikler:
        df2.drop([i], axis=1, inplace=True)
df2.head()
    
```

### 1. Model

```python
Y= df2["hedef"]
X= df2[["Humidity", "Wind Speed (km/h)"]]
X= sm.add_constant(X)
sonuclar = sm.OLS(Y,X).fit()
sonuclar.summary()
```

### 2. Model

```python
#  yukarıdaki modele nem ve rüzgar hızı etkileşimini dahil edin ve OLS'yi kullanarak modeli tahmin edin.
df2["nem_ruzgar_hızı"]= df2["Humidity"]*df2["Wind Speed (km/h)"]
Y= df2["hedef"]
X= df2[["Humidity", "Wind Speed (km/h)","nem_ruzgar_hızı"]]
X= sm.add_constant(X)
sonuclar= sm.OLS(Y,X).fit()
sonuclar.summary()
```

### 3. Model

```python
# İlk modele ek açıklayıcı değişken olarak görünürlük ekleyin ve tahmin edin.
df2["Visibility (km)"]= df["Visibility (km)"]
Y= df2["hedef"]
X= df2[["Humidity", "Wind Speed (km/h)", "nem_ruzgar_hızı","Visibility (km)"]]
X= sm.add_constant(X)
sonuclar= sm.OLS(Y,X).fit()
sonuclar.summary()
```

```python
df= pd.read_csv("Desktop/Bootcamp Ödevler/house_prices.csv")
df.head()
```

```python
ozellikler=  ["SalePrice","LotArea", "YearBuilt", "BldgType", "Heating", "GarageQual"]
df2= df.copy()
for i in df.columns:
    if i not in ozellikler:
        df2.drop([i], axis=1, inplace=True)
df2.head()
```

```python
dummies= ["GarageQual","Heating","BldgType"]
df3= pd.concat([df2, pd.get_dummies(df2[dummies])], axis=1)
df3.drop(dummies, axis=1, inplace=True)
df3.head()
```

### 1. Model

```python
acıklayıcılar= []
for i in df3.columns:
    if i != "SalePrice":
        acıklayıcılar.append(i)
```

```python
Y= df3["SalePrice"]
X= df3[acıklayıcılar]
X= sm.add_constant(X)
sonuclar= sm.OLS(Y,X).fit()
sonuclar.summary()
```

### 2. Model

```python
df3["TotRmsAbvGrd"]= df["TotRmsAbvGrd"]
df3.head()
```

```python
df= pd.read_csv("Desktop/Bootcamp Ödevler/weatherHistory.csv")
df.head()
```

```python
df["hedef"]= df["Temperature (C)"]-df["Apparent Temperature (C)"]
# açıklayıcılar = Humidity ve  Wind Speed (km/h)
```

```python
Y = df.hedef
X= df[["Humidity", "Wind Speed (km/h)"]]

X= sm.add_constant(X)
sonuclar = sm.OLS(Y,X).fit()
sonuclar.summary()
```

```python
Y = df.hedef
df["nem_ruzgar"]= df["Humidity"]*df["Wind Speed (km/h)"]
X= df[["Humidity", "Wind Speed (km/h)", "nem_ruzgar"]]
X= sm.add_constant(X)
sonuc= sm.OLS(Y,X).fit()
sonuc.summary()
```

```python
df= pd.read_csv("Desktop/Bootcamp Ödevler/house_prices.csv")
df.head()
```

```python
df2 = df.copy()  
df2.head()
```

```python
kolonlar_2 = ["SalePrice","LotArea", "Condition1", "YearBuilt", "BldgType", "OverallCond",
              "Heating", "TotRmsAbvGrd", "GarageQual", "PoolArea"]

for kolon in df.columns:
    if kolon not in kolonlar_2:
        df2.drop([kolon], axis=1, inplace=True)
df2.columns       
```

```python
object_kolonlar= ["Condition1", "BldgType", "Heating", "GarageQual"]
for kolonlar in object_kolonlar:
    df2[kolonlar]= df2[kolonlar].astype("category")
df2.info()
```

```python
category_kolonlar= ["Condition1", "BldgType", "Heating", "GarageQual"]
df3= pd.concat([df2, pd.get_dummies(df[category_kolonlar])], axis=1)
df3.drop(category_kolonlar, axis=1, inplace=True)
df3.head()
```

```python
butun_kolonlar= df3.columns
istenen_kolonlar= []
for i in butun_kolonlar:
    if i != "SalePrice":
        istenen_kolonlar.append(i)
istenen_kolonlar
        
```

```python
Y= df3["SalePrice"]
X= df3[istenen_kolonlar]
X = sm.add_constant(X)
sonuclar= sm.OLS(Y,X).fit()
sonuclar.summary()
```

```python
istenmeyenler= ["GarageQual_TA","GarageQual_Fa"]
yeni_ozellikler = []
for i in istenen_kolonlar:
    if i not in istenmeyenler:
        yeni_ozellikler.append(i)
yeni_ozellikler
```

```python
Y= df3["SalePrice"]
X= df3[yeni_ozellikler]
X= sm.add_constant(X)
sonuclar= sm.OLS(Y,X).fit()
sonuclar.summary()
 
```
